```python
# Identifying additional columns as nulls 
data = pd.read_csv(r"C:\Users\Hp\OneDrive\Documents\retail orders project\orders.csv",
                   na_values=["Not Available","unknown"])
```

```python
data["Ship Mode"].isnull().sum()
```

```python
data["Ship Mode"].unique()
```

```python
data_cleaned = data.dropna(subset=['Ship Mode'])
```

```python
len(data_cleaned)
```

```python
# transforming uppercase columns to lowercase while replacing whitespaces with underscores
data_cleaned.columns = data_cleaned.columns.str.lower()
data_cleaned.columns = data_cleaned.columns.str.replace(" ","_")
```

```python
data_cleaned.head()
```

```python
# adding useful columns giving additional insights
data_cleaned["discount"] = data_cleaned["list_price"]*data_cleaned["discount_percent"]*.01
data_cleaned["sale_price"] = data_cleaned["list_price"]-data_cleaned["discount"]
data_cleaned["profit"] = data_cleaned["sale_price"]-data_cleaned["cost_price"]
```

```python
data_cleaned.head()
```

```python
# dropping unneeded columns
data_cleaned = data_cleaned.drop(["cost_price","list_price","discount_percent"],axis=1)
```

```python
data_cleaned.head()
```

```python
print(data_cleaned["order_date"].dtype)
```

```python
data_cleaned["order_date"] = pd.to_datetime(data_cleaned["order_date"])
```

```python
print(data_cleaned["order_date"].dtype)
```

```python
#saving the modified dataframe now to a new csv file
data_cleaned.to_csv("modified_orders.csv",index=False)
```
